## Factors and Multiples

Factors: All numbers that divides a number N completely are called its factors

Example: Factors of 6 are :- 1, 2, 3, 6

Multiples: A number N is a multiple of all its divisiors/factors.

Example: Multiples of 6 are :- 6, 12, 18, .....

### 1. Given a number N, print all its divisors

#### Naive solution

##### Approach:

Traverse numbers from 1 to N and print its divisors

```python
def printDivisors(N):
    print("Divisors of {} are: ".format(N))
    for i in range(1, N+1):
        if N % i == 0:
            print(i, end=" ")

N = int(input("Enter a number: "))
printDivisors(N)
```

Time Complexity: O(N)

Space Complexity: O(1)

#### Efficient solution

##### Approach:

Traverse till the square root of a number N

```python
def printDivisors(N):
    print("Divisors of {} are: ".format(N))
    i = 1
    while i*i < N:
        if N % i == 0:
            print(i, end=" ")
        i += 1
    x = i
    while i >= 1:
        if N % i == 0 and N//i >= x:
            print(N//i, end=" ")
        i -= 1
        
N = int(input("Enter a number: "))
printDivisors(N)
```

Time Complexity: O(Sqrt(N))
    
Space Complexity: O(1)
