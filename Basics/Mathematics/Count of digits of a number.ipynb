## Count digits of a number

### Naive solution

1. #### Iterative solution

```python
def countDigits(N):
    ans = 0
    while N > 0:
        ans += 1
        N //= 10
    return ans
N = int(input("Enter a number: "))
print("Digits count of {} is: {}".format(N, countDigits(N)))
```

Time Complexity: O(log n)

Space complexity: O(1)

2. #### Recursive solution

```python
def countDigits(N):
    if N <= 9:
        return 1
    return 1 + countDigits(N//10)
N = int(input("Enter a number: "))
print("Digits count of {} is: {}".format(N, countDigits(N)))
```

Time Complexity: O(log n)

Space Complexity: O(n)

3. #### Logarthamic solution

```python
import math
def countDigits(N):
    return int(math.log(N,10)) + 1
N = int(input("Enter a number: "))
print("Digits count of {} is: {}".format(N, countDigits(N)))
```

Time Complexity: O(1)
    
Space Complexity: O(1)
