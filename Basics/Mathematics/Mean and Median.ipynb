## Mean and Median

Mean - Sum of n numbers divided by n

Median - Middle element (or) sum of 2 middle elements after sorting a list of numbers

Median = f(x) = {

                    ((n/2)th element + (n/2-1)th element / 2) if n is Even
                    
                    n/2 th element     if n is Odd
                    
                }
