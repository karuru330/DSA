## HCF and LCM

HCF : Highest common factor of 2 or more numbers
    
Example: HCF of (15,12) is 3. 3 is the highest number which divides both 15 and 12
    
LCM : Least common multiple of 2 or more numbers

Example: LCM of (3,4) is 12. 12 is the leastcommon multiple of both 3 and 4

### 1. Given 2 numbers, find HCF

#### Naive solution

##### Approach:

1. Find max of 2 numbers
2. Find the number which divides both numbers by iterating from max number till 1.

```python
def gcd(a,b):
    m = max(a,b)
    while a%m != 0 or b%m != 0:
        m -= 1
    return m

a = int(input("Enter number1: "))
b = int(input("Enter number2: "))
print("GCD of {} and {} is {}".format(a,b,gcd(a,b)))
```

Time Complexity: O(max(a,b))
    
Space Complexity: O(1)

#### Efficient solution

##### Approach:

Approach is to use Euclidean algorithm to find GCD


```python
def gcd(a,b):
    if b == 0:
        return a
    return gcd(b, a%b)

a = int(input("Enter number1: "))
b = int(input("Enter number2: "))
print("GCD of {} and {} is {}".format(a,b,gcd(a,b)))
```

Time Complexity: O(log(a,b))
    
Space complexity: O(1)

### 2. Given 2 numbers, find LCM

#### Naive solution

##### Approach:

1. Find Max of 2 numbers
2. Find the least multiple of both numbers using a loop

```python
def lcm(a,b):
    m = max(a,b)
    while m%a != 0 or m%b != 0:
        m += 1
    return m

a = int(input("Enter number1: "))
b = int(input("Enter number2: "))
print("LCM of {} and {} is {}".format(a,b,lcm(a,b)))
```

Time Complexity: O(a*b - max(a,b))
    
Space Complexity: O(1)

#### Efficient solution

##### Approach:

LCM(a,b) = (a*b)/GCD(a,b)

```python
def gcd(a,b):
    if b == 0:
        return a
    return gcd(b, a%b)

def lcm(a,b):
    return (a*b)//gcd(a,b)

a = int(input("Enter number1: "))
b = int(input("Enter number2: "))
print("LCM of {} and {} is {}".format(a,b,lcm(a,b)))
```

Time Complexity: O(log(a,b))

Space Complexity: O(1)
