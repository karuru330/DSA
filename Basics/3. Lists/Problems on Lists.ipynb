## Problems on Lists

### 1. Given a list of elements, find 2nd largest element

#### Naive solution

#### Approach:

1. Find largest element
2. Find second largest element in a separate loop

```python
def find2ndMax(L):
    m = L[0]
    for i in range(1, len(L)):
        if m < L[i]:
            m = L[i]
            
    sm = -999999
    for i in range(1, len(L)):
        if sm < L[i] and L[i] < m:
            sm = L[i]
    return sm

L = list(map(int, input("Enter a list of numbers separated by a space: ").split()))
print("2nd largest element of the List is: {}".format(find2ndMax(L)))
```

Time Complexity: Theta(N)
    
Space Complexity: O(1)

#### Efficient solution

#### Approach:

1. use 2 variables to store max and second max
2. use single loop instead of 2 loops

```python
def find2ndMax(L):
    m = L[0]
    sm = -99999999
    for i in range(1, len(L)):
        if m < L[i]:
            sm = m
            m = L[i]
    return sm

L = list(map(int, input("Enter a list of numbers separated by a space: ").split()))
print("2nd largest element of the List is: {}".format(find2ndMax(L)))
```

Time Complexity: Theta(N) -- Reduces one loop from the above naive solution

Space Complexity: O(1)

### 2. Given a list of elements, reverse it

#### Naive solution

##### Approach:

1. Traverse the list from right to left, append each element to a resultant list
2. Return the resultant list

```python
def reverse(L):
    res = []
    for i in range(len(L)-1, -1, -1):
        res.append(L[i])
    return res

L = list(map(int, input("Enter a list of numbers separated by a space: ").split()))
print("After reversing the list is: {}".format(reverse(L)))
```

Time Complexity: O(N)
    
Space Complexity: O(N)

#### Efficient solution

##### Approach:

1. Approach here to use two pointers method
2. use 2 variables to point beg, end of the List
3. swap 1st and last element.
4. move the pointers 1 step forward and backward accordingly and do the swapping

```python
def reverse(L):
    i = 0
    j = len(L) - 1
    while i < len(L)//2:
        L[i], L[j] = L[j], L[i]
        i += 1
        j -= 1
    return L

L = list(map(int, input("Enter a list of numbers separated by a space: ").split()))
print("After reversing the list is: {}".format(reverse(L)))
```

Time Complexity: O(N)

Space Complexity: O(1)

### 3. Given a sorted list of elements, remove duplicates from it

#### Naive solution

##### Approach:

1. Iterate through each element of the list, store unique elements in a temp list
2. copy all the elements from temp list to original list


```python
def removeDuplicates(L):
    temp = [0]*len(L)
    temp[0] = L[0]
    res = 1
    for i in range(1, len(L)):
        if temp[res-1] != L[i]:
            temp[res] = L[i]
            res += 1
    for i in range(res):
        L[i] = temp[i]
    return L, res


L = list(map(int, input("Enter a list of numbers separated by a space: ").split()))
L, res = removeDuplicates(L)
print("After removing duplicates, the list is: {}".format(L[:res]))  
```

Time Complexity: O(N)
    
Space Complexity: O(N)

#### Efficient solution

##### Approach:

Approach here is to use 2-pointer approach to remove duplicates without using extra space

```python
def removeDuplicates(L):
    res = 1
    for i in range(1, len(L)):
        if L[res-1] != L[i]:
            L[res] = L[i]
            res += 1
        
    return L, res

L = list(map(int, input("Enter a list of numbers separated by a space: ").split()))
L, res = removeDuplicates(L)
print("After removing duplicates, the list is: {}".format(L[:res])) 
```

Time Complexity: O(N)

Space Complexity: O(1)

### 4. Given a list of elements, rotate the list by 1 element

#### Direct solutions:

##### Using Slicing

```python
def rotateList(L):
    return L[1:]+L[0:1]

L = list(map(int, input("Enter a list of numbers separated by a space: ").split()))
print("After rotation, the list is: {}".format(rotateList(L))) 
```

##### Using pop() and append()

```python
def rotateList(L):
    L.append(L.pop(0))
    return L

L = list(map(int, input("Enter a list of numbers separated by a space: ").split()))
print("After rotation, the list is: {}".format(rotateList(L)))
```

##### Own solution

```python
def rotateList(L):
    x = L[0]
    for i in range(1, len(L)):
        L[i-1] = L[i]
    L[-1] = x
    return L

L = list(map(int, input("Enter a list of numbers separated by a space: ").split()))
print("After rotation, the list is: {}".format(rotateList(L)))
```

Time Complexity: O(N)

Space Complexity: O(1)
