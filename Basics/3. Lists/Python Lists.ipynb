## Lists in Python

List is a dynamic sized container which allows to store values of different data type

Allows random access

#### Basic operations :

```python
L = [1,2,3,4,5]

print(L)

print(L[-1]) # prints last element

print(L[::-1]) # prints list in a reverse order

L.append(10) # adds 10 at the end

L.insert(3, 2) # inserts 100 at index 3

L.count(2) # counts occurence of 2 in the list

L.index(3) # finds index of element 30

print(5 in L) # checks 5 is present in L

print(L.index(3, 0, len(L))) # finds index of element 3 in the index range 0 - len(L)

L.pop() # removes last element

L.remove(5) # removes element 5

del L[1] # removes element at index 1

L.pop(2) # removes element at index 2

print(max(L)) # prints max element in the list

print(min(L)) # prints min element in the list

L.reverse() # reverses the list

L.sort() # sorts the list in-place

print(sum(L)) # prints sum of elements in the list
```

### Slicing

L[start:stop:skip]

```python
L = [1,2,3,4,5,6]

print(L[0:4:2]) # prints [1,3]
```
