## Problems on binary search

### 1. Given a list of sorted elements and a number x, find the first occurence of number x in the list

I/P: arr[] = {1, 10, 10, 10, 20, 20, 40}
       
       x = 20
       
O/P: 4


I/P: arr[] = {10, 20, 30}

     x = 15
     
O/P: -1

#### Approach:

1. Use binary search to find the element
2. If any element found, check if the element is present in the left side and return the min index

```python
def firstOccurence(l, x):
    start = 0
    end = len(l) - 1
    while start <= end:
        mid = (start + end)//2
        if l[mid] == x:
            if l[mid] == 0 or l[mid] != l[mid-1]:
                return mid
            end = mid - 1
        elif l[mid] < x:
            start = mid + 1
        else:
            end = mid - 1
    return -1

l = list(map(int, input("Enter a list of sorted elements with space separated:\n").split()))
x = int(input("Enter an element to search: "))
print("Element", x, end=" ")
ind = firstOccurence(l, x)
if ind != -1:
    print("found at index {}".format(ind))
else:
    print("not found")
```

Time Complexity: O(log N)

Space Complexity: O(1)

### 2. Given a list of sorted elements and a number x, find the last occurence of number x in the list

I/P: arr[] = {1, 10, 10, 10, 20, 20, 40}
       
       x = 20
       
O/P: 5


I/P: arr[] = {10, 20, 30}

     x = 15
     
O/P: -1

#### Approach:

1. Use binary search to find the element
2. If any element found, check if the element is present in the right side and return the max index

```python
def firstOccurence(l, x):
    start = 0
    end = len(l) - 1
    while start <= end:
        mid = (start + end)//2
        if l[mid] == x:
            if l[mid] == len(l)-1 or l[mid] != l[mid + 1]:
                return mid
            start = mid + 1
        elif l[mid] < x:
            start = mid + 1
        else:
            end = mid - 1
    return -1

l = list(map(int, input("Enter a list of sorted elements with space separated:\n").split()))
x = int(input("Enter an element to search: "))
print("Element", x, end=" ")
ind = firstOccurence(l, x)
if ind != -1:
    print("found at index {}".format(ind))
else:
    print("not found")
```

Time Complexity: O(log N)

Space Complexity: O(1)

### 3. Given a list of sorted elements and a number x, find the occurence of number x in the list

I/P: arr[] = {1, 10, 10, 10, 20, 20, 40}
       
       x = 20
       
O/P: 2


I/P: arr[] = {10, 20, 30}

     x = 15
     
O/P: -1

#### Approach:

1. Find 1st occurence
2. Find last occurence
3. Last occurence - first occurence + 1 will be the answer

```python
def findFirstOccurence(L, x):
    start = 0
    end = len(L)-1
    while start <= end:
        mid = (start + end)//2
        if L[mid] == x:
            if mid == 0 or L[mid-1] != x :
                return mid
            end = mid - 1
        elif L[mid] > x:
            end = mid - 1
        else:
            start = mid + 1
    return -1

def findLastOccurence(L, x):
    start = 0
    end = len(L)-1
    while start <= end:
        mid = (start + end)//2
        if L[mid] == x:
            if mid == len(L)-1 or L[mid + 1] != x :
                return mid
            start = mid + 1
        elif L[mid] > x:
            end = mid - 1
        else:
            start = mid + 1
    return -1

l = list(map(int, input("Enter a list of sorted elements with space separated:\n").split()))
x = int(input("Enter an element to search: "))
print("Element", x, end=" ")
ind = findFirstOccurence(l,x)
if ind != -1:
    ind = findLastOccurence(l, x) - ind  + 1
    print("Occurent count: {}".format(ind))
else:
    print("not found")
```

Time Complexity: O(Log N)
    
Space Complexity: O(1)

### 4. Given a list of 0's and 1's in sorted order, find the first occurence of 1's list

I/P: arr[] = {0, 0, 0, 1, 1, 1, 1}
              
O/P: 4


I/P: arr[] = {0, 0, 0}
     
O/P: 0

#### Naive Solution:

##### Approach:

1. Simple search

```python
def count1s(L):
    ans = 0
    for x in L:
        if x == 1:
            ans += 1
    return ans

l = list(map(int, input("Enter a list of 0's and 1's in sorted order with space separated:\n").split()))
count = count1s(l)
print("Count of 1s: {}".format(count))
```

Time Complexity: O(N)
    
Space Complexity: O(1)

#### Efficient solution:

##### Approach:

1. Find first occurence of 1
2. Length of list - first occurence

```python
def firstOccurence(L, x):
    start = 0
    end = len(L)-1
    while start <= end:
        mid = (start + end)//2
        if L[mid] == x:
            if mid == 0 or L[mid] != L[mid-1]:
                return mid
            end = mid - 1
        elif L[mid] > x:
            end = mid - 1
        else:
            start = mid + 1
    return 0

l = list(map(int, input("Enter a list of 0's and 1's in sorted order with space separated:\n").split()))
count = firstOccurence(l, 1)
if count == 0:
    count = 0
else:
    count = len(l) - count
print("Count of 1s: {}".format(count))
```

Time Complexity: O(Log N)
    
Space Complexity: O(1)

### 5. Given a number N, find the square root of it.

I/P: N = 15

O/P: 3

I/P: 16

O/P: 4


#### Naive solution:

##### Approach:

Iterate from i = 1 till i*i <= N

```python
def sqrt(N):
    i = 1
    while i*i <= N:
        i += 1
    return i-1

N = int(input("Enter a number: "))
print("Square root of {} is {}".format(N, sqrt(N)))
```

Time Complexity: O(Sqrt(N))

Space Complexity: O(1)

#### Efficient solution:

##### Approach:

Approach here is to use binary search to reduce the no.of iterations

```python
def sqrt(N):
    start = 1
    end = N
    ans = -1
    while start <= end:
        mid = (start + end)//2
        msq = mid * mid
        if msq == N:
            return mid
        elif msq < N:
            start = mid + 1
            ans = mid
        else:
            end = mid - 1
    return ans
            
N = int(input("Enter a number: "))
print("Square root of {} is {}".format(N, sqrt(N)))
```

Time Complexity: O(Sqrt(N))

Space Complexity: O(1)

### 6. Floor in a Sorted Array

Given a sorted array arr[] of size N without duplicates, and given a value x. Floor of x is defined as the largest element K in arr[] such that K is smaller than or equal to x. Find the index of K(0-based indexing).

I/P: arr[] = {1, 2, 8, 9, 10, 11, 12, 19}
     X = 5
     
O/P: 1

#### Naive solution:

##### Approach:

Iterate the array and find the answer

```python
def findFloor(A,N,X):
        #Your code here
        ans = -1
        K = -1
        for i in range(N):
            if A[i] < X and A[i] > K:
                ans = i
                K = A[i]
        return ans
    
A = list(map(int, input("Enter list of sorted elements separated by space:\n").split()))
X = int(input("Enter a number: "))
ans = findFloor(A, len(A), X)
if ans != -1:
    print("Floor of {} found at index {}".format(X, ans))
else:
    print("Floor of {} not present in the array".format(X))
```

Time Complexity: O(N)

Space Complexity: O(1)

#### Efficent solution:

##### Approach:

Use binary search to solve this problem

```python
def findFloor(A,N,X):
        #Your code here
        start = 0
        end = N-1
        K = -1
        ans = -1
        while start <= end:
            mid = (start + end)//2
            if A[mid] <= X and A[mid] > K:
                ans = mid
                K = A[mid]
            elif A[mid] > X:
                end = mid - 1
            else:
                start = mid + 1
        return ans
    
A = list(map(int, input("Enter list of sorted elements separated by space:\n").split()))
X = int(input("Enter a number: "))
ans = findFloor(A, len(A), X)
if ans != -1:
    print("Floor of {} found at index {}".format(X, ans))
else:
    print("Floor of {} not present in the array".format(X))
```

Time Complexity: O(log N)

Space Complexity: O(1)

### 7. Peak Element

An element is called a peak element if its value is not smaller than the value of its adjacent elements(if they exists). Given an array arr[] of size N, Return the index of any one of its peak elements. 
Note: The generated output will always be 1 if the index that you return is correct. Otherwise output will be 0. Also, arr[] will be in ascending order before the peak element, and in descending order after that.

I/P: arr[] = {1,2,3}

O/P: 2

#### Naive solution:

##### Approach:

Iterate the array and compare each element with its adjacent elements

```python
def findPeak(arr):
    if len(arr) == 1:
        return 0
    for i in range(len(arr)):
        if i == 0 and arr[i] >= arr[i+1]:
            return i
        elif i == len(arr)-1 and arr[i-1] <= arr[i]:
            return i
        elif arr[i] >= arr[i-1] and arr[i] >= arr[i+1]:
            return i
    return -1

A = list(map(int, input("Enter list of sorted elements separated by space:\n").split()))
ans = findPeak(A)
print("Peak element ", end="")
if ans != -1:
    print("found at index {}".format(ans))
else:
    print("not found")
```

Time Complexity: O(N)

Space Complexity: O(1)

#### Efficent solution:

##### Approach:

Use binary search

If there is a greater element at left side, go to left and search for peak element
otherwise go to right and find the peak element

```python
def peakElement(arr):
        # Code here
        n = len(arr)
        if n == 1:
            return 0
        start = 0
        end = n-1
        while start <= end:
            mid = (start + end)//2
            if mid == 0:
                if arr[mid + 1] <= arr[mid]:
                    return mid
                start = mid + 1
            elif mid == n-1:
                if arr[mid - 1] <= arr[mid]:
                    return mid
                end = mid - 1
            else:
                if arr[mid] >= arr[mid-1] and arr[mid] >= arr[mid+1]:
                    return mid
                elif arr[mid] < arr[mid-1]:
                    end = mid - 1
                else:
                    start = mid + 1
        return -1
    
A = list(map(int, input("Enter list of sorted elements separated by space:\n").split()))
ans = peakElement(A)
print("Peak element ", end="")
if ans != -1:
    print("found at index {}".format(ans))
else:
    print("not found")
```

Time Complexity: O(log N)
    
Space Complexity: O(1)
