## Binary search in python

Binary search algorithm is used to search an element in sorted array

Time Complexity: O(log N)

### Binary Search

Input: l = [10, 20, 30, 40, 50, 60]

       x = 20
       
Output: 1

```python
def binarySearch(l, x):
    start = 0
    end = len(l)
    while start <= end:
        mid = (start + end)//2
        if l[mid] == x:
            return mid
        elif l[mid] < x:
            start = mid + 1
        else:
            end = mid - 1
    return -1

l = list(map(int, input("Enter a list of sorted elements with space separated:\n").split()))
x = int(input("Enter an element to search: "))
print("Element", x, end=" ")
ind = binarySearch(l, x)
if ind != -1:
    print("found at index {}".format(ind))
else:
    print("not found")
```

Time Complexity: O(log N)

Space Complexity: O(1)

### Recursive Binary search

```python
def recursiveBinarySearch(l, x, start, end):
    if start <= end:
        mid = (start + end)//2
        if l[mid] == x:
            return mid
        elif l[mid] < x:
            return recursiveBinarySearch(l, x, mid + 1, end)
        else:
            return recursiveBinarySearch(l, x, start, mid - 1)
    return -1

l = list(map(int, input("Enter a list of sorted elements with space separated:\n").split()))
x = int(input("Enter an element to search: "))
print("Element", x, end=" ")
ind = recursiveBinarySearch(l, x, 0, len(l)-1)
if ind != -1:
    print("found at index {}".format(ind))
else:
    print("not found")
```

Time Complexity: O(log N)

Space Complexity: O(log N)
