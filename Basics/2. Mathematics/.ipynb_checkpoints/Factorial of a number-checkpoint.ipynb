## Factorial of a number

### 1. Find Factorial of a Number

#### Iterative solution

```python
def factorial(N):
    fact = 1
    while N > 0:
        fact *= N
        N -= 1
    return fact

N = int(input("Enter a number: "))
print("Factorial of {} is {}".format(N, factorial(N)))
    
```

Time Complexity: O(N)

Space Complexity: O(1)

#### Recursive solution

```python
def factorial(N):
    if N <= 0:
        return 1
    return N * factorial(N-1)

N = int(input("Enter a number: "))
print("Factorial of {} is {}".format(N, factorial(N)))
```

Time Complexity: O(N)

Space Complexity: O(N)

### 2. Given a number, find no.of digits in its factorial

#### Naive Solution

##### Approach: 

1. Find Factorial of number N
2. Find the digits of factorial using Logarthamic solution

```python
import math

def factorial(N):
    fact = 1
    while N > 0:
        fact *= N
        N -= 1
    return fact

def findDigits(N):
    return math.floor(math.log(factorial(N), 10)) + 1

N = int(input("Enter a number: "))
print("Count of digits of {}! is {}".format(N, findDigits(N)))
```

Time Complexity: O(N+d) where d is number of digits

Space Complexity: O(1)

#### Efficient solution

##### Approach:

log(a*b) = log(a) + log(b)

log(N!) = log(1 x 2 x 3 x.....x N)

log(1 x 2 x 3 x ......x N) = log(1) + log(2) + log(3) +.........+ log(N)

```python
import math
def findDigits(N):
    ans = 0
    while N >= 1:
        ans += math.log(N,10)
        N -= 1
    return math.floor(ans) + 1

N = int(input("Enter a number: "))
print("count of digits of {}! is {}".format(N, findDigits(N)))
```

Time Complexity: O(n logn)
    
Space Complexity: O(1)

### 3. Given a number, find no.of trailing zeros in its factorial

#### Naive solution

##### Approach:

1. Find factorial
2. Find trailing zeros using a loop

```python
def factorial(N):
    fact = 1
    while N > 0:
        fact *= N
        N -= 1
    return fact

def findTrailingZeros(N):
    fact = factorial(N)
    ans = 0
    while fact%10 == 0:
        ans += 1
        fact //= 10
    return ans

N = int(input("Enter a number: "))
print("No of trailing zeros in {}! is {}".format(N, findTrailingZeros(N)))
```

Time Complexity: O(N+d) where d is number of digits

Space Complexity: O(1)

#### Efficient solution

##### Approach:

Here the approach is to use prime factorization

5! can be written as (2 x 2 x 2 x 3 x 5)

A trailing zero is always formed with 2 and 5

It is observed that the count of 2's is always greater than count of 5's in the prime factorization of a number

Hence, if we find count of 5's in in the prime factorization of a number, our job is done

To find no.of 5's, we can divide the number N by 5^1, 5^2, and so on

```python
def findTrailingZeors(N):
    k = 1
    ans = 0
    while 5**k <= N:
        ans += (N//(5**k))
        k += 1
    return ans

N = int(input("Enter a number: "))
print("Count of trailing zeros of {}! is {}".format(N, findTrailingZeros(N)))
```

Time Complexity: O(log5 n)

Space Complexity: O(1)
