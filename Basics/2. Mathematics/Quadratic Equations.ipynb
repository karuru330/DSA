## Quadratic Equations

Quadratic equations are in the form of ax^2 + bx + c

There can be 2 roots at most of a quardatic equation

Roots are depends on Discriminant D = sqrt(b^2-4ac)

if D < 0 - Imaginary roots

D > 0 - Distinct roots

D == 0 - Equal roots

Root can be calculated by using below formula

(-b + sqrt(b^2-4ac)) / 2a

(-b - sqrt(b^2-4ac)) / 2a
