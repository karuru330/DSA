## Power of a number

Power of a,b is given as a^b

### 1. Given two numbers a and b. Find a^b

#### Naive solution

##### Approach:

Iterate loop b times, multiply a to the resultant variable each time

```python
def power(a,b):
    res = 1
    for i in range(1, b+1):
        res *= a
    return res

a = int(input("Enter number a: "))
b = int(input("Enter number b: "))
print("{}^{} is {}".format(a, b, power(a,b)))
```

Time Complexity: O(b)

Space Complexity: O(1)

#### Efficient solution

##### Approach:

Every number can be represented as power of 2

3^10 = 3^8 + 3^2

Binary representation of 10: 1010

So, traverse the binary representation of value b from LSB to MSB. 

When the bit is set, multiply the value of a to resultant variable

Else multiply a to itself

```python
def power(a,b):
    res = 1
    while b > 0:
        if b%2 != 0:
            res = res * a
        a = a*a
        b //= 2
    return res

a = int(input("Enter number a: "))
b = int(input("Enter number b: "))
print("{}^{} is {}".format(a, b, power(a,b)))
```

Time Complexity: O(log b)

Space Complexity: O(1)
