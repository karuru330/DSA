## Insertion Sort

- In-place
- Stable
- Best case: O(N)
- Worst case: O(N^2)


Approach:

1. Starting from 2nd element, compare it with adjacent elements and swap it till the next left element is not greater

```python
def insertionSort(L):
    n = len(L)
    for i in range(1, n):
        j = i
        element = L[j]
        while j > 0 and L[j] < L[j-1]:
            L[j], L[j-1] = L[j-1], L[j]
            j -= 1
L = list(map(int,input("Enter list of elements with space separated:\n").split()))
print("Before sorting List is", L)
insertionSort(L)
print("After sorting List is", L)
```

Time Complexity: O(N^2)

Space Complexity: O(1)
