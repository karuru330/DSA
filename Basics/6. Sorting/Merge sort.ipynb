## Merge Sort

- Stable
- Not in-place
- Uses O(N) extra space
- Uses Divide and conquer approach

```python
def merge(L, beg, mid, end):
    temp = []
    i = beg
    j = mid + 1
    while i <= mid and j <= end:
        if L[i] < L[j]:
            temp.append(L[i])
            i += 1
        else:
            temp.append(L[j])
            j += 1

    while i <= mid:
        temp.append(L[i])
        i += 1

    while j <= end:
        temp.append(L[j])
        j += 1

    i = beg
    k = 0
    while k < len(temp):
        L[i] = temp[k]
        k += 1
        i += 1
def mergeSort(L, beg, end):
    if beg < end:
        mid = (beg + end)//2
        mergeSort(L, beg, mid)
        mergeSort(L, mid+1, end)
        merge(L, beg, mid, end)

L = list(map(int,input("Enter list of elements with space separated:\n").split()))
print("Before sorting List is", L)
mergeSort(L, 0, len(L)-1)
print("After sorting List is", L)
```

Time Complexity: O(N log N)

Space Complexity: O(N)
