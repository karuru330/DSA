## Selection Sort

- InPlace
- Not stable

Approach:

1. In each iteration, find ith min element and put it at ith location

```python
def selectionSort(L):
    n = len(L)
    for i in range(n-1):
        min_index = i
        for j in range(i+1, n):
            if L[j] < L[min_index]:
                min_index = j
        L[i], L[min_index] = L[min_index], L[i]

L = list(map(int,input("Enter list of elements with space separated:\n").split()))
print("Before sorting List is", L)
selectionSort(L)
print("After sorting List is", L)
```

Time Complexity: Theta(N^2)

Space Complexity: O(1)
