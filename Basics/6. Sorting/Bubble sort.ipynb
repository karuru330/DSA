## Bubble Sort

Approach:

1. Compare each adjacent elements and swap
2. ith max element will be moved to the right side of the list/array in each pass

```python
def bubbleSort(L):
    n = len(L)
    for i in range(n-1):
        for j in range(n-i-1):
            if L[j] > L[j+1]:
                L[j], L[j+1] = L[j+1], L[j]
    
L = list(map(int,input("Enter list of elements with space separated:\n").split()))
print("Before sorting List is", L)
bubbleSort(L)
print("After sorting List is", L)
```

Time Complexity: O(N^2)

Space Complexity: O(1)

#### Optimized bubble sort

Approach:

1. Use a flag to find any swaping is done in the ith iteration
2. If no swap is done in any iteration, stop the loop execution

```python
def bubbleSort(L):
    n = len(L)
    for i in range(n-1):
        sorted = False
        for j in range(n-i-1):
            if L[j] > L[j+1]:
                L[j], L[j+1] = L[j+1], L[j]
                sorted = True
        if sorted == False:
            break
    
L = list(map(int,input("Enter list of elements with space separated:\n").split()))
print("Before sorting List is", L)
bubbleSort(L)
print("After sorting List is", L)
```
