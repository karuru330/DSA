## Sorting in Python

### sort()

- Inplace
- works only for List

### sorted()

- works for any iterator
- sorts and returns new iterator

Both of the methods uses Tim sort algorithm(Hybrid algorith - uses merge sort and insertion sort internally.)

Worst case time complexity is O(N log N)

#### Example 1: Sort a list using sort() method

```python
L = [8, 1, 99, 76, 34, 55]
L.sort()
print(L)
```

#### Example 2: Sort a list using sorted() method

```python
L = [8, 1, 99, 76, 34, 55]
L = sorted(L)
print(L)
```

#### Example 3: Sort a list in reverse order

```python
L = [8, 1, 99, 76, 34, 55]
L.sort(reverse=True)
print(L)
```

#### Example 4: Sort user defined objects

```python
def compare(obj):
    return obj.x

class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

L = [Point(1, 8), Point(5, 9), Point(2, 67), Point(1, 17)]
L.sort(key=compare)
for obj in L:
    print(obj.x, obj.y)
```

##### Method 2: Using lt() method

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    
    def __lt__(self, other):
        if self.x == self.x:
            return self.y < other.y
        return self.x < other.x

L = [Point(1, 8), Point(5, 9), Point(2, 67), Point(1, 17)]
L.sort(key=compare)
for obj in L:
    print(obj.x, obj.y)
```

#### Stability In Sorting:

A sorting algorithm is called stable if it ensures the order of the items while sorting as they appreared in the original array/list

##### Stable sorting algorithms: Bubble sort, Merge sort, Insertion sort

##### UnStable sorting algorithms: Selection sort, Quick sort, Heap sort
