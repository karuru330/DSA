## Recursion

A function that calls itself directly or indirectly

Base case - It is a condition which decides the end of the recursive function

Recursive call - Calling the function itself with updated/new parameters

##### Applications:
    
1. Many algorithmic techniques are based on the recursion

    a. Divide and conquer

    b. Dynamic programming
    
    c. Backtracking
    
    
2. Tower of Hanoi

3. DFS

##### Tail Recursion

A recursive function is called tail recursive, when it does not do anything after the last recursive call

###### Example of Tail recursion

```python
def fun(N):
    if N <= 1:
        return
    print("Hello Dilli")
    fun(N-1)

fun(5)
```

##### Example of Non-Tail recursion

```python
def fun(N):
    if N <= 1:
        return 
    fun(N-1)
    print("Hello Dilli")
    
fun(5)
```
