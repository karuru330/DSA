## Problems on Recursion

### 1. Check whether a string is a palindrome

```python
def isPalindrome(string, start, end):
    if start <= end:
        isPalindrome(string, start+1, end-1)
        return string[start] == string[end]
    
string = input("Enter a string: ")
print("{} is ".format(string), end="")
if isPalindrome(string, 0, len(string)-1):
    print("a palindrome")
else:
    print("not a palindrome")
```

Time Complexity: O(N)
    
Space Complexity: O(N)

### 2. Print Nth Fibonacci number

```python
def fib(N):
    if N <= 2:
        return 1
    return fib(N-1) + fib(N-2)

N = int(input("Enter a number: "))
print("{}th Fibonacci number is: {}".format(N, fib(N)))        
```

Time Complexity: O(2^N)
    
Space Complexity: O(N)

### 3. Find GCD using recursion

```python
def gcd(a,b):
    if b == 0:
        return a
    return gcd(b, a%b)

a = int(input("Enter a number: "))
b = int(input("Enter a number: "))
print("GCD of ({},{}) is {}".format(a, b, gcd(a,b)))
```

Time Complexity: O(log N)

Space Complexity: O(log N)
